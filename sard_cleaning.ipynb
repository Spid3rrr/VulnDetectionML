```python
def clean_source_code(file_content):
    import re

    ## Delete comments from file_content
    # Delete multiline comments
    COMMENT_START = "<!--"
    COMMENT_END = "-->"
    comment_starts = [m.start() for m in re.finditer(COMMENT_START, file_content)]
    comment_ends = [m.start() for m in re.finditer(COMMENT_END, file_content)]
    for i in range(len(comment_starts)):
        file_content = file_content[:comment_starts[i]] + file_content[comment_ends[i]+len(COMMENT_END):]
    # Delete lines starting with #
    file_content = "\n".join([line for line in file_content.split("\n") if not line.startswith("#")])
    # Delete empty lines
    file_content = "\n".join([line for line in file_content.split("\n") if line.strip() != ""])
    
    ## Find all variable names in file_content and rename them
    # Find all variable names in a php file
    variable_names = re.findall(r"\$[a-zA-Z_][a-zA-Z0-9_]*", file_content)
    # Remove duplicates
    variable_names = list(set(variable_names))
    variable_names.sort(key = len, reverse=True)
    # Rename all variable names in file_content
    for variable_name in variable_names:
        new_variable_name = "$variable" + str(variable_names.index(variable_name))
        file_content = file_content.replace(variable_name, new_variable_name)
    
    ## Rename all functions in file_content
    # Find all function names in a php file
    function_names = re.findall(r"function [a-zA-Z_][a-zA-Z0-9_]*", file_content)
    # Remove duplicates
    function_names = list(set(function_names))
    # Remove "function " from function names
    function_names = [function_name.replace("function ", "") for function_name in function_names]
    # Rename all function names in file_content
    for function_name in function_names:
        new_function_name = "function" + str(function_names.index(function_name))
        file_content = file_content.replace(function_name, new_function_name)
    ## Rename all classes in file_content
    # Find all class names
    class_names = re.findall(r"class [a-zA-Z_][a-zA-Z0-9_]*", file_content)
    # Remove duplicates
    class_names = list(set(class_names))
    # Remove "class " from class names
    class_names = [class_name.replace("class ", "") for class_name in class_names]
    # Rename all class names in file_content
    for class_name in class_names:
        new_class_name = "Class" + str(class_names.index(class_name))
        file_content = file_content.replace(class_name, new_class_name)
    
    return file_content
```
