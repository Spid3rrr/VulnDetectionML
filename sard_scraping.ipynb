```python
import requests
url = 'https://samate.nist.gov/SARD/api/test-cases/search?language%5B%5D=php&page=1&limit=100'
response = requests.get(url)
response = response.json()
print(response)
pageCount = response['pageCount']
print(pageCount)
```

```python
# extract source code from test case
import zipfile
import os

# create folder to store source code
folder_path = "dataset_full"
zip_path = "dataset_full_zip"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
if not os.path.exists(zip_path):
    os.makedirs(zip_path)


def get_source_code(case):
    file_url = case["download"]
    file = requests.get(file_url)
    filename = file_url.split("/")[-1]
    src_file = case["sarif"]["runs"][0]["artifacts"][0]["location"]["uri"]
    is_good = case["sarif"]["runs"][0]["properties"]["state"] == "good"
    vuln_type = (
        case["sarif"]["runs"][0]["taxonomies"][0]["name"]
        + case["sarif"]["runs"][0]["taxonomies"][0]["taxa"][0]["id"]
    )
    new_filename = (
        case["identifier"] + "_" + vuln_type + "_" + ("0" if is_good else "1") + ".txt"
    )
    with open(zip_path + "/" + filename, "wb") as f:
        f.write(file.content)
    with zipfile.ZipFile(zip_path + "/" + filename, "r") as zip_ref:
        with zip_ref.open(src_file) as src:
            src_code = src.read()
            f = open(folder_path + "/" + new_filename, "wb")
            f.write(src_code)
            f.close()
    if os.path.isfile(zip_path + "/" + filename):
        os.remove(zip_path + "/" + filename)
```

```python
case = response['testCases'][0]
get_source_code(case)
```

```python
import _thread
for i in range(pageCount-1):
    print('page: ', i+1, '/', pageCount)
    url = 'https://samate.nist.gov/SARD/api/test-cases/search?language%5B%5D=php&page='+str(i+1)+'&limit=100'
    response = requests.get(url)
    response = response.json()
    for case in response['testCases']:
        _thread.start_new_thread(get_source_code, (case, ))
```
